```python
%matplotlib inline
import pandas as pd 
import matplotlib.pyplot as plt #python plotting library. plotting framework
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_csv('cluster2.csv')
print('Dimensions of the data = {}'.format(df.shape))
```

```python
X = df.values # values in csv file
cmap = 'tab10' # Store the name of colormap in global variable cmap
```

```python
# Kmeans
for i in range(1,6):
    k_means = KMeans(n_clusters=i, random_state=0)
    y_pred = k_means.fit_predict(X)
    plt.figure(figsize=(5,5))
    # c = color of eacj points, cmap = set the color map
    plt.scatter(X[:,0],X[:,1], c= y_pred, s=4, cmap=cmap)
plt.show()
```

```python
# DBSCAN
for i, (eps,min_samples) in enumerate(((0.03,6),(0.04,6),(0.05,6),(0.06,20),(0.06,15),(0.06,6),(0.06,3))):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    y_pred = dbscan.fit_predict(X)
    plt.figure(i+1, figsize=(5,5))
    plt.scatter(X[:,0],X[:,1], cmap=cmap, s=4, c = y_pred)
    plt.title("eps: {}, min_samples: {}".format(eps, min_samples))
plt.show()
```
