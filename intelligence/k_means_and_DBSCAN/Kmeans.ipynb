```python
%matplotlib inline
import pandas as pd 
import matplotlib.pyplot as plt #python plotting library. plotting framework
```

```python
df = pd.read_csv('cluster2.csv')
print('Dimensions of the data = {}'.format(df.shape))
```

```python
df[:5] # csv file
```

```python
X = df.values # values in csv file
X[:5]
```

```python
# set the size of the figure
plt.figure(figsize = (5,5)) # width, height in inches
# plot the data points
# performs a scatter plot(x,y coordinates, size of each points)
plt.scatter(X[:,0],X[:,1], s= 4) 
plt.show()
```

```python
cmap = 'tab10' # Store the name of colormap in global variable cmap
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=4, random_state=0)
```

```python
y_pred = k_means.fit_predict(X)
print(y_pred[:10])
print(k_means.cluster_centers_)
```

```python
plt.figure(figsize=(5,5))
# c = color of eacj points, cmap = set the color map
plt.scatter(X[:,0],X[:,1], c= y_pred, s=4, cmap=cmap)
plt.show()
```
